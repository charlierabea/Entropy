```python
import cv2
import os
import util
import gabor
import matplotlib.pyplot as plt
import numpy as np
from pathlib import Path


# 設置圖像目錄
parent_dir = r"/home/charlie/projects/entropy/raw/Raw_superficial/"

#第一層 列出reactive/treatment
for dirname in os.listdir(parent_dir):
    if (dirname.endswith(".DS_Store")):
        continue
    
    cond_dir = parent_dir + dirname+'/'
    print(cond_dir)
    
    #第二層 列出active/inactive
    for dirname2 in os.listdir(cond_dir):
        if (dirname2.endswith(".DS_Store")):
            continue
            
        amd_dir = cond_dir + dirname2+'/'
        print(amd_dir)
        
        #列出目錄中的所有圖像
        for filename in os.listdir(amd_dir):
            if not (filename.endswith(".bmp") or filename.endswith(".jpg")):
                continue
                
            # 組合文件路徑
            filepath = os.path.join(amd_dir, filename)
            #print(filepath)
            
            # 讀取圖像
            img = cv2.imread(filepath, cv2.IMREAD_GRAYSCALE)
            img_gabor = cv2.imread(filepath)
            
            print(filename)
            print(img.shape)
            #plt.imshow(img)
            
            binaryprocesses = [('otsu', util.do_otsu),
                                ('mean', util.do_mean),
                                ('gaussmean', util.do_gaussmean),
                                ('triangle', util.do_triangle)]
            
            charprocesses = [('faz', util.do_faz),
                                ('edge', util.do_canny),]
            
            for have_gabor_filter in {True, False}:
                #宣告img1
                img1 = img.copy()
                
                if have_gabor_filter:
                    #存
                    imgdir_path = f'/home/charlie/projects/entropy/preprocessed/gabor/{dirname}/{dirname2}/'
            
                    try:
                        os.makedirs(imgdir_path)
                    # 檔案已存在的例外處理
                    except FileExistsError:
                        print("資料夾已存在。")
            
                    img_path = imgdir_path + filename
                    img_path2 = Path(imgdir_path + filename)
                    if img_path2.exists():
                        print('檔案已存在')
                        continue
                    
                    img1 = gabor.Gabor_process(img_gabor)
                    cv2.imwrite(img_path,  img1)
                    
                #存faz/canny
                for charprocess_name, charprocess in charprocesses:
                    #宣告資料夾名稱
                    mypath = ""
                    if have_gabor_filter:
                        mypath += "gabor_"
                        
                    #進行處理
                    #
                    mypath += charprocess_name
                    
                    #製造資料夾
                    imgdir_path = f'/home/charlie/projects/entropy/preprocessed/{mypath}/{dirname}/{dirname2}/'
                    try:
                        os.makedirs(imgdir_path)
                    # 資料夾已存在的例外處理
                    except FileExistsError:
                        print("資料夾已存在。")

                    #檔案存檔
                    img_path = imgdir_path + filename
                    img_path2 = Path(imgdir_path + filename)
                    if img_path2.exists():
                        print('檔案已存在')
                        continue
                    
                    img_char = charprocess(img1)
                    cv2.imwrite(img_path,  img_char)
                
                #存binary
                for binaryprocess_name, binaryprocess in binaryprocesses:
                    #宣告資料夾名稱
                    mypath = ""
                    if have_gabor_filter:
                        mypath += "gabor_"
                    
                    #進行處理
                    #
                    mypath += binaryprocess_name

                    #製造資料夾
                    imgdir_path = f'/home/charlie/projects/entropy/preprocessed/{mypath}/{dirname}/{dirname2}/'
                    try:
                        os.makedirs(imgdir_path)
                    # 資料夾已存在的例外處理
                    except FileExistsError:
                        print("資料夾已存在。")
                    
                    #檔案存檔
                    img_path = imgdir_path + filename
                    img_path2 = Path(imgdir_path + filename)
                    if img_path2.exists():
                        print('檔案已存在')
                        continue
                    
                    img_binary = binaryprocess(img1)
                    img_binary = util.do_fillholes(img_binary)
                    cv2.imwrite(img_path,  img_binary)
            
```
