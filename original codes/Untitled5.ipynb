```python
plt.figure(num = 'OCTa' ,figsize=(14, 12))


img = cv2.imread(r'C:\Users\vghuser\Desktop\test_1ppp.png')
#print(type(img))
plt.subplot(2, 3, 1)
plt.title('orginal')
plt.imshow(img)


orginal_thresh = cv2.threshold(img, 45, 255, cv2.THRESH_BINARY)[1]
plt.subplot(2, 3, 2)
plt.title('thresh')
plt.imshow(orginal_thresh)


#img_bilater = cv2.bilateralFilter(img,9,75,75)
#img_median = cv2.medianBlur(img, 5)


img_blur = cv2.blur(img, (5,5))
blur_thresh = cv2.threshold(img_blur, 45, 255, cv2.THRESH_BINARY)[1]
plt.subplot(2, 3, 3)
plt.title('blur')
plt.imshow(blur_thresh)


img_guassian = cv2.GaussianBlur(img, (5,5),0)
guassian_thresh = cv2.threshold(img_guassian, 45, 255, cv2.THRESH_BINARY)[1]
plt.subplot(2, 3, 4)
plt.title('guassian')
plt.imshow(guassian_thresh)


kernel = np.array(([ 0.0625, 0.125, 0.0625 ], [ 0.125, 0.25, 0.125 ], [ 0.0625, 0.125, 0.0625]))#, dtype= " float32 " )
img_2d = cv2.filter2D(img, -1 , kernel)
thresh = cv2.threshold(img_2d, 45, 255, cv2.THRESH_BINARY)[1]
#thresh = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)[1]
#thresh = cv2.threshold(img, 50, 255, cv2.THRESH_BINARY)
#thresh = cv2.adaptiveThreshold(img, 0, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2)[1]
plt.subplot(2, 3, 5)
plt.title('2D_thresh')
plt.imshow(thresh)


#edge_detection
img_gray = cv2.imread(r'C:\Users\vghuser\Desktop\test_1ppp.png', cv2.IMREAD_GRAYSCALE)
blur_gray = cv2.GaussianBlur(img_gray, (5,5),0)
edges_canny = cv2.Canny(blur_gray, 10, 200)
plt.subplot(2, 3, 6)
plt.title('edges')
plt.imshow(edges_canny)

plt.savefig("C:/Users/vghuser/Desktop/OCTa_test_process_5.png")
plt.show()
```
