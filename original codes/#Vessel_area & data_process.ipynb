```python

import cv2
import os

# 設置圖像目錄
image_dir = r"/home/charlie/projects/entropy/all/active/"

# 初始化總面積
results = {}

# 列出目錄中的所有文件
for filename in os.listdir(image_dir):
    # 檢查文件是否為圖像
    if not (filename.endswith(".bmp") or filename.endswith(".png")):
        continue

    # 組合文件路徑
    filepath = os.path.join(image_dir, filename)

    # 讀取圖像
    img = cv2.imread(filepath, cv2.IMREAD_GRAYSCALE)
    
    # 進行二值化
    _, img = cv2.threshold(img, 128, 255, cv2.THRESH_BINARY)
    
    # 轉換為灰度圖
    #gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    # 使用 Canny 邊緣檢測算法檢測血管
    edges = cv2.Canny(img, 50, 150)
    
    # 找到血管
    contours, _ = cv2.findContours(edges, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)

    # 計算連通區域面積
    area = 0
    for c in contours:
        area += cv2.contourArea(c)

    results[filename] = area

# print or save the results
print(results)
#or 
with open("F:/vessels_result_Canny.csv","w") as f:
    for k,v in results.items():
        f.write(f"{k}: {v}\n")
```

```python
#自適應二值化

import cv2
import os

# 設置圖像目錄
image_dir = r"F:\results_vessel_200\rose\second_stage_v2\noThresh/"

# 初始化總面積
results = {}

# 列出目錄中的所有文件
for filename in os.listdir(image_dir):
    # 檢查文件是否為圖像
    if not (filename.endswith(".bmp") or filename.endswith(".png")):
        continue

    # 組合文件路徑
    filepath = os.path.join(image_dir, filename)

    # 讀取圖像
    img = cv2.imread(filepath, cv2.IMREAD_GRAYSCALE)

    # 進行自適應二值化
    img = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 0)

    # 尋找連通區域
    contours, _ = cv2.findContours(img, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # 計算連通區域面積
    area = 0
    for c in contours:
        area += cv2.contourArea(c)

    results[filename] = area

# print or save the results
#print(results)
#or 
with open("F:/vessels_result_contours_adaptive.csv","w") as f:
    for k,v in results.items():
        f.write(f"{k}: {v}\n")
```
