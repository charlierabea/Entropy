```python
import cv2
import os
import util
import gabor
import matplotlib.pyplot as plt

# 設置圖像目錄
parent_dir = r"/home/charlie/projects/entropy/raw/Raw_superficial/"

reactive_dir = os.path.join(parent_dir, "reactive/")
treatment_dir = os.path.join(parent_dir, "treatment/")


#第一層 列出reactive/treatment
for dirname in os.listdir(parent_dir):
    if (dirname.endswith(".DS_Store")):
        continue
    
    cond_dir = parent_dir + dirname+'/'
    print(cond_dir)
    
    #第二層 列出active/inactive
    for dirname2 in os.listdir(cond_dir):
        if (dirname2.endswith(".DS_Store")):
            continue
            
        amd_dir = cond_dir + dirname2+'/'
        print(amd_dir)
        
        #列出目錄中的所有圖像
        for filename in os.listdir(amd_dir):
            if not (filename.endswith(".bmp") or filename.endswith(".jpg")):
                continue
                
            # 組合文件路徑
            filepath = os.path.join(amd_dir, filename)
            #print(filepath)
            
            # 讀取圖像
            img = cv2.imread(filepath, cv2.IMREAD_GRAYSCALE)
            
            print(img.shape)
            #plt.imshow(img)
            imgdir_path = f'/home/charlie/projects/entropy/img/{dirname}/{dirname2}/'
            
            try:
                os.makedirs(imgdir_path)
            # 檔案已存在的例外處理
            except FileExistsError:
                print("檔案已存在。")
            
            img_path = imgdir_path + filename
            cv2.imwrite(img_path,  img)
            
            
            i=1
            while(i>0):
                #gabor
                
                if(i==2):
                    img_gabor = gabor.Gabor_process(img)
                    #進行高斯模糊
                    img_gabor_gaussian = util.do_GaussianBlur(img_gabor)
                    print(filename)
                    cv2.imwrite("/home/charlie/projects/entropy/img/gabor_gaussian/"+img_path, img_gabor_gaussian)
                    
                    
                    # 進行二值化（otsu/mean/gaussian_mean)
                    j=3
                    while(j>0):
                        #otsu
                        if(j==3):
                            img_gabor_gaussian_otsu = util.do_otsu(img_gabor_gaussian)
                            cv2.imwrite("/home/charlie/projects/entropy/img/gabor_gaussian_otsu/"+img_path, img_gabor_gaussian_otsu)
                            j=j-1
                        
                        #mean
                        elif(j==2):
                            img_gabor_gaussian_mean = util.do_mean(img_gabor_gaussian)
                            cv2.imwrite("/home/charlie/projects/entropy/img/gabor_gaussian_mean/"+img_path, img_gabor_gaussian_mean)
                            j=j-1
                        
                        #gaussian_mean
                        elif(j==1):
                            img_gabor_gaussian_mean = util.do_gaussmean(img_gabor_gaussian)
                            cv2.imwrite("/home/charlie/projects/entropy/img/gabor_gaussian_gaussmean/"+img_path, img_gabor_gaussian_gaussmean)
                            j=j-1
                    i= i-1
                
                #no_gabor
                if(i==1):
                    #img_gabor = gabor.Gabor_process(img)
                    #進行高斯模糊
                    #img_gaussian = img
                    
                    #cv2.imwrite("/home/charlie/projects/entropy/img/gabor_gaussian/"+img_path, img_gabor_gaussian)
                    
                    # 進行二值化（otsu/mean/gaussian_mean)
                    j=3
                    while(j>0):
                        #otsu
                        if(j==3):
                            img_otsu = util.do_otsu(img)
                            #plt.imshow(img_otsu)
                            cv2.imwrite("/home/charlie/projects/entropy/img/otsu/"+img_path, img_otsu)
                            
                            j=j-1
                        
                        #mean
                        elif(j==2):
                            img_mean = util.do_mean(img)
                            cv2.imwrite("/home/charlie/projects/entropy/img/mean/"+img_path, img_mean)
                            j=j-1
                        
                        #gaussian_mean
                        elif(j==1):
                            img_gaussmean = util.do_gaussmean(img)
                            cv2.imwrite("/home/charlie/projects/entropy/img/gaussmean/"+img_path, img_gaussmean)
                            j=j-1
                    i= i-1
                            
                    
                    


            
            
            
```
